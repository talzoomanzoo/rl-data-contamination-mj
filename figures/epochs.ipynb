```python
import matplotlib.pyplot as plt

# Example data (replace with your actual values)
epochs = [0, 1, 2]
accuracy_run1 = [0.62, 0.68, 0.71]
accuracy_run2 = [0.70, 0.75, 0.80]

# Create the plot
plt.figure()
plt.plot(epochs, accuracy_run1, marker='o', label='Member x2')
plt.plot(epochs, accuracy_run2, marker='o', label='Member x3')

# Labels and title
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')

# Ensure x-axis shows only 0, 1, 2
plt.xticks([0, 1, 2])

# Optional: set limits
plt.ylim(0, 1)

# Grid for readability
plt.grid()
plt.legend()

# Show plot
plt.show()
```
