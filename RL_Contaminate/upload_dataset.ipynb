```python
import os
from datasets import Dataset
from huggingface_hub import HfApi, create_repo

# Path to local parquet file
PARQUET_PATH = "/workspace/RL-Data-Contamination/benchmarks/KK/RLMIA_kk.parquet"

# Target dataset repo (org/name or user/name)
HF_DATASET_REPO_ID = "talzoomanzoo/rlmia_kk"


# Set to True if you want a private dataset repo
PRIVATE = False

api = HfApi()
create_repo(repo_id=HF_DATASET_REPO_ID, private=PRIVATE, exist_ok=True, repo_type="dataset")
# If the repo already exists and is private, make it public to avoid private storage limits.
api.update_repo_visibility(repo_id=HF_DATASET_REPO_ID, private=PRIVATE, repo_type="dataset")

print(f"Uploading {PARQUET_PATH} to https://huggingface.co/datasets/{HF_DATASET_REPO_ID} ...")

dataset = Dataset.from_parquet(PARQUET_PATH)
dataset.push_to_hub(HF_DATASET_REPO_ID)
print("Upload complete.")
```
